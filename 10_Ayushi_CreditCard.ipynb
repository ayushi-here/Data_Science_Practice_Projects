**Response Variable Analysis**

```python
class_name={0:'Not Fraud',1:'Fraud'}
df.Class.value_counts().rename(index=class_name)
# Fraud vs Non-fraud
```

```python
from sklearn.model_selection import train_test_split
y=df['Class']
x=df.drop('Class',axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(y.isnull())
```

**Data modeling**

```python
df = df.dropna(subset=['Class'])
```

```python
print(y.isnull())
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
df.fillna(0,inplace=True)
```

**Logistic Regression Classifier**

```python
logisreg= LogisticRegression()
logisreg.fit(x_train,y_train)
y_pred=logisreg.predict(x_test)
acc_logisreg=round(accuracy_score(y_pred,y_test)*100,2)
print("Accuracy of Logistic Regression: ",acc_logisreg)
```

**Linear Discriminant Analysis**

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
model=LinearDiscriminantAnalysis()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
acc_ida=round(accuracy_score(y_pred,y_test)*100,2)
print("Accuracy of Linear Discriminant Analysis: ",acc_ida)
```

**Gaussian Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
acc_gnb=round(accuracy_score(y_pred,y_test)*100,2)
print("Accuracy of Gaussian Naive Bayes: ",acc_gnb)
```

**Dicision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
acc_dt=round(accuracy_score(y_pred,y_test)*100,2)
print("Accuracy of Decision Tree Classifier: ",acc_dt)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
acc_rf=round(accuracy_score(y_pred,y_test)*100,2)
print("Accuracy of Random Forest: ",acc_rf)
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
acc_knn=round(accuracy_score(y_pred,y_test)*100,2)
print("Accuracy of KNN: ",acc_knn)
```

```python
models=pd.DataFrame({'Model':['Logistic Regression','Linear Discriminant Analysis','Gaussian Naive Bayes','Decision Tree Classifier','Random Forest','KNN'],'Accuracy':[acc_logisreg,acc_ida,acc_gnb,acc_dt,acc_rf,acc_knn]})
```
